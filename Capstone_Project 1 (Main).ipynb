##  3. Feature Engineering

### - Task 10: Create new features that might be useful for the analysis (e.g., date-related features from timestamps, interaction terms).

```python
if 'Order_Date' in df_original.columns:
  # Convert 'Order_Date' to datetime objects
  df_original['Order_Date'] = pd.to_datetime(df_original['Order_Date'])
  # Extract date-related features
  df['Order_Year'] = df_original['Order_Date'].dt.year
  df['Order_Month'] = df_original['Order_Date'].dt.month
  df['Order_Day'] = df_original['Order_Date'].dt.day
  df['Order_Day_of_Week'] = df_original['Order_Date'].dt.dayofweek # Monday=0, Sunday=6
  df['Order_Week_of_Year'] = df_original['Order_Date'].dt.isocalendar().week.astype(int)

  print("\nAdded date-related features:")
  print(df[['Order_Year', 'Order_Month', 'Order_Day', 'Order_Day_of_Week', 'Order_Week_of_Year']].head())
```

```python
# Interaction between Cost and Discount
df['Cost_x_Discount'] = df['Cost_of_the_Product'] * df['Discount_offered']

# Interaction between Weight and Customer Rating
df['Weight_x_Rating'] = df['Weight_in_gms'] * df['Customer_rating']
```

```python
print("\nAdded interaction features:")
print(df[['Cost_x_Discount', 'Weight_x_Rating']].head())
```

```python
print("\nDataFrame shape after adding new features:", df.shape)
```

#### Summary
- We extracted new time-related features from the order date, such as day of the week and week of the year, which may influence delivery speed. We also created interaction terms to capture compounded effects (e.g., high-cost items with high discounts). These engineered features aim to enhance the predictive power of our model.

## 4. Model Building

### - Task 12: Split the dataset into training and testing sets.

```python
from sklearn.model_selection import train_test_split
```

```python
if 'Reached.on.Time_Y.N' in df.columns:
    X = df.drop('Reached.on.Time_Y.N', axis=1)
    y = df['Reached.on.Time_Y.N']
    print("\nFeatures (X) and Target (y) defined.")
    print("Shape of X:", X.shape)
    print("Shape of y:", y.shape)
 
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

    print("\nDataset split into training and testing sets.")
    print("Shape of X_train:", X_train.shape)
    print("Shape of X_test:", X_test.shape)
    print("Shape of y_train:", y_train.shape)
    print("Shape of y_test:", y_test.shape)
    print("\nProportion of target classes in original dataset:")
    print(y.value_counts(normalize=True))
    print("\nProportion of target classes in training set:")
    print(y_train.value_counts(normalize=True))
    print("\nProportion of target classes in testing set:")
    print(y_test.value_counts(normalize=True))

else:
    print("\nTarget column 'Reached.on.Time_Y.N' not found in the DataFrame.")
    print("Please identify and define the target variable before splitting the dataset.")
```

 #### Summary
 - I split the dataset into training and testing sets using `train_test_split` from `sklearn.model_selection`.
 - Features (X) are all columns except the target variable ('Reached.on.Time_Y.N').
 - The target variable (y) is 'Reached.on.Time_Y.N'.
 - A `test_size` of 0.25 means 25% of the data is allocated to the test set and 75% to the training set.
 - `random_state=42` ensures that the split is the same every time the code is run, making the results reproducible.
 - `stratify=y` is used to maintain the same proportion of the target variable classes in both the training and testing sets, which is important for classification problems, especially with imbalanced datasets.

### - Task 13: Train a simple linear regression model (if the task is regression) or a logistic regression model (if the task is classification).

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print("\nTraining a Logistic Regression model...")

model = LogisticRegression(random_state=42, solver='liblinear', class_weight='balanced') # 'liblinear' solver is good for small datasets
```

```python
# Train the model using the training data
model.fit(X_train, y_train)
print("Logistic Regression model trained successfully.")
```

```python
# Predict on the testing data
y_pred = model.predict(X_test)

print("\nModel evaluation on the testing set:")
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
class_report
```

#### The Logistic Regression model achieved an accuracy of ~66%, which is relatively low. This may be due to the models inability to capture non-linear relationships in the data, class imbalance, or feature complexity. In subsequent steps, we explore more robust models like Random Forests and KNN, and focus on additional metrics like F1-score and ROC-AUC to better evaluate classification performance.

#### Summary
- We initialized and trained a Logistic Regression model on the training data (`X_train`, `y_train`).
- Logistic Regression is appropriate because the task is classification (predicting a binary outcome: 0 or 1).
- After training, we used the model to predict the target variable on the test set (`X_test`).
- We evaluated the model's performance using standard metrics:
  - **Accuracy:** The proportion of correctly classified instances.
  - **Confusion Matrix:** A table summarizing the performance of a classification algorithm. Each row represents the instances in an actual class while each column represents the instances in a predicted class.
  - **Classification Report:** Provides precision, recall, F1-score, and support for each class.

## 5. Model Tuning

### - Task 16: Perform hyperparameter tuning using GridSearchCV or RandomizedSearchCV.

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
print("\n--- Performing Hyperparameter Tuning ---")
```

```python
# --- Hyperparameter Tuning for Random Forest ---
print("\n--- Tuning Random Forest Classifier using GridSearchCV ---")

param_grid_rf = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'class_weight': ['balanced']
}
```

```python
grid_search_rf = GridSearchCV(estimator=RandomForestClassifier(random_state=42),
                              param_grid=param_grid_rf,
                              cv=5,
                              scoring='f1_weighted',
                              n_jobs=-1,
                              verbose=1)
```

```python
# Fit GridSearchCV to the training data
grid_search_rf.fit(X_train, y_train)

print("\nBest parameters found for Random Forest:")
print(grid_search_rf.best_params_)

print("\nBest weighted F1-score on training set for Random Forest:")
print(grid_search_rf.best_score_)
```

```python
# Evaluate the best Random Forest model on the test set
best_rf_model = grid_search_rf.best_estimator_
y_pred_best_rf = best_rf_model.predict(X_test)

print("\nEvaluation of Best Random Forest model on the testing set:")
accuracy_best_rf = accuracy_score(y_test, y_pred_best_rf)
f1_weighted_best_rf = f1_score(y_test, y_pred_best_rf, average='weighted')
try:
    roc_auc_best_rf = roc_auc_score(y_test, best_rf_model.predict_proba(X_test)[:, 1])
except AttributeError:
    roc_auc_best_rf = "N/A (Probabilities not available)"

print(f"Best Random Forest Accuracy: {accuracy_best_rf:.4f}")
print(f"Best Random Forest Weighted F1-score: {f1_weighted_best_rf:.4f}")
print(f"Best Random Forest ROC AUC: {roc_auc_best_rf}")
print("\nBest Random Forest Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_best_rf))
print("\nBest Random Forest Classification Report:")
print(classification_report(y_test, y_pred_best_rf))
```

```python
# --- Hyperparameter Tuning for KNN (Example using RandomizedSearchCV) ---
print("\n--- Tuning KNN Classifier using RandomizedSearchCV ---")

param_dist_knn = {
    'n_neighbors': np.arange(1, 31), # Try k from 1 to 30
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
```

```python
random_search_knn = RandomizedSearchCV(estimator=KNeighborsClassifier(),
param_distributions=param_dist_knn,
 n_iter=20, # Number of parameter combinations to try
cv=5,
scoring='f1_weighted',
n_jobs=-1,
random_state=42,
verbose=1)                                       
```

```python
# Fit RandomizedSearchCV to the training data
random_search_knn.fit(X_train, y_train)

print("\nBest parameters found for KNN:")
print(random_search_knn.best_params_)

print("\nBest weighted F1-score on training set for KNN:")
print(random_search_knn.best_score_)
```

```python
# Evaluate the best KNN model on the test set
best_knn_model = random_search_knn.best_estimator_
y_pred_best_knn = best_knn_model.predict(X_test)

print("\nEvaluation of Best KNN model on the testing set:")
accuracy_best_knn = accuracy_score(y_test, y_pred_best_knn)
f1_weighted_best_knn = f1_score(y_test, y_pred_best_knn, average='weighted')
try:
    roc_auc_best_knn = roc_auc_score(y_test, best_knn_model.predict_proba(X_test)[:, 1])
except AttributeError:
    roc_auc_best_knn = "N/A (Probabilities not available)"

print(f"Best KNN Accuracy: {accuracy_best_knn:.4f}")
print(f"Best KNN Weighted F1-score: {f1_weighted_best_knn:.4f}")
print(f"Best KNN ROC AUC: {roc_auc_best_knn}")
print("\nBest KNN Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_best_knn))
print("\nBest KNN Classification Report:")
print(classification_report(y_test, y_pred_best_knn))
```

#### Summary
- I performed hyperparameter tuning for the Random Forest and KNN models.
- For Random Forest, I used GridSearchCV to systematically search through a defined grid of hyperparameter combinations.
- For KNN, I used RandomizedSearchCV, which samples a fixed number of parameter settings from specified distributions. Randomized search can be faster when the search space is large.
